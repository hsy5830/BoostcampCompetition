# Over Sampling

## up to 2000

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
# import random
from numpy import random

df = pd.read_csv('data/train.csv')
df.head()
df['label'].value_counts()
```

```python
df_over = df
label_dist = df['label'].value_counts()

# over sampling
np.random.seed(42)
for label in df['label'].unique():
    if label_dist[label] <= 2000:    # elif 추가로 기존 개수에 따라 차등 복사 가능
        idx = random.choice(list(df.loc[df['label']==label].id), int(2000-label_dist[label])) # 2000개가 되도록 복사
        df_over = pd.concat([df_over, df.iloc[idx, :]], axis = 0)
print(df_over.shape)
        
# shuffling data
df_over = df_over.sample(frac=1)  # row 전체 shuffle
df_over = df_over.sample(frac=1).reset_index(drop=True)  # shuffling하고 index reset
df_over
```

```python
df_over.to_csv('/Users/youp/Desktop/train_oversampling2000.csv', encoding="utf-8-sig", index=False)
```
