## preprocess

* elastic search 안에 preprocess 함수

```python
import re

def preprocess(text):
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"\\n", " ", text)
    text = re.sub(r"#", " ", text)
    text = re.sub(r"[^A-Za-z0-9가-힣.?!,()~‘’“”"":%&《》〈〉''㈜·\-\'+\s一-龥サマーン]", "", text)  # サマーン 는 predictions.json에 있었음
    text = re.sub(r"\s+", " ", text).strip()  # 두 개 이상의 연속된 공백을 하나로 치환
    # text = re.sub(r"[^A-Za-z0-9가-힣.?!,()~‘’“”"":%&《》〈〉''㈜·\-\'+\s一-龥]", "", text)
    
    return text
```

```python
wiki_sen_pre = list(map(preprocess, wiki_sen))
wiki_sen_pre
```

## klue/mrc in Wiki?

* 대회의 testset은 klue/mrc의 train, validation에서 일부를 가져온 것
* 확인해보기

```python
# 각 셋에서 중복되는 context 제거
train_set = list(set(df_klue_train['context']))
valid_set = list(set(df_klue_valid['context']))

print(f"train : 중복된 {len(df_klue_train) - len(train_set)} 개 제거 -> 총 {len(train_set)} 개")
print(f"valid : 중복된 {len(df_klue_valid) - len(valid_set)} 개 제거 -> 총 {len(valid_set)} 개")
```

```python
# unique한 context 기준
c1, c2 = 0, 0
for i in range(len(train_set)): # 초
    for j in range(len(wiki_sen)):
        if train_set[i] in wiki_sen[j]: c1 += 1
for i in range(len(valid_set)): # 10초
    for j in range(len(wiki_sen)):
        if valid_set[i] in wiki_sen[j]: c2 += 1
        
print(f"train의 {len(train_set)} 중 {c1}개가 wiki에 포함")
print(f"valid의 {len(valid_set)} 중 {c2}개가 wiki에 포함")
```
