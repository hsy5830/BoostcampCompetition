# def ensemble()

key, value = text, prob 이용

```python
def ensemble(dict_list, answer_key, ans, topk=1):
    N = len(dict_list) # model 개수
    
    l_dict = []
    for i in range(len(dict_list[0])): # ~600
        tmp_dict = {}
        for j in range(N):
            tmp_val = dict_list[j][answer_key[i]]
            for k in range(20):
#                 print(i ,j, k)
                if tmp_val[k]['text'] in tmp_dict.keys():
                    tmp_dict[tmp_val[k]['text']] += tmp_val[k]['probability']
                else:
                    tmp_dict[tmp_val[k]['text']] = tmp_val[k]['probability']
        l_dict.append(tmp_dict)
        
    l_id = []
    for i in range(len(l_dict)): # 600
        tmp = sorted(l_dict[i].items(), key = lambda x: x[1], reverse = True)[0:1]
        ids = [x[0] for x in tmp] # id만 가져오기 -> text임
        l_id.append(ids)
    
    answer = {key:val[0] for key,val in zip(answer_key, l_id)}

    c =0
#     ans = [str(i) for i in list(test_true['answers'])]
#     ans = [eval(i)['text'][0] for i in ans]
    for i in range(len(answer)):
        if answer[answer_key[i]] in eval(test_true['answers'][i])['text']: c+=1
    
    print(f"Accuracy: {c}/{len(test_true)} = {c/len(test_true)}")
    return answer
```

```python
ensemble(df_list[11:13], answer_key, ans)
```

# 앙상블 !!!!

```python
N = len(file_list) # 모형 개수
N
```

```python
results = []
for i in range(1,N+1):
    print(f"---------------------------------------- {N}개 중 {i}개 선택 ----------------------------------------")
    comb = list(itertools.combinations(range(N), i))

    for j in range(len(comb)):
        ldf = []
        for k in range(len(comb[j])):
            ldf.append(copy.deepcopy(df_list[comb[j][k]]))
        ensemble(ldf, answer_key, ans)
        print(f"           사용한 데이터 인덱스 : {comb[j]}")

    print("\n")
```
